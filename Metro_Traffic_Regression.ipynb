```python
# Regression: Drought Prediction

# IMPORTANT NOTE:
# When trying to execute the regression in R, I got an error in R, stating: "vector memoy exhausted (limit reached?)"
# Even after following instructions on StackOverflow and some guesswork, I am still not able to eliminate this error.
# Hence, this code is written in python instead of R. I hope this is not a problem.

# Import necessary libraries
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm

metro = pd.read_csv("/content/Metro_Interstate_Traffic_Volume.csv")

# Split the date_time column into two columns date and time
metro[['date','time']] = metro.date_time.str.split(" ",expand=True)


# Split the date column into the columns year, month, and day
metro[['year','month', 'day']] = metro.date.str.split("-",expand=True)
metro['year'] = metro['year'].astype(int)


# Linear Regression: First approach using all parameters
model = smf.ols('traffic_volume ~ year + month + day + time + temp + rain_1h + snow_1h + clouds_all + weather_main', metro).fit()
print(model.summary())
print(sm.stats.anova_lm(model, typ = 2))


# Linear Regression: Second Approach, eliminate all parameters with p-value > 0.05

# Drop weather description values first
dropped_weather = ['Haze', 'Mist', 'Rain', 'Squall']
indexWeather = metro[metro['weather_main'].isin(dropped_weather)].index
metro.drop(indexWeather , inplace=True)

model2 = smf.ols('traffic_volume ~ month + time + temp + clouds_all + weather_main', metro).fit()
print(model2.summary())
print(sm.stats.anova_lm(model2, typ = 2))
```
